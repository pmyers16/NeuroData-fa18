```python
import os
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

from load_emb import emb
```

```python
# Collect all embeddings
def collect_all_embs():
    path = '/Users/vivek/Documents/GitHub/NDD/test/struc2vec/emb/emb_10d/'
    filelist = [file for file in os.listdir(path) if '.emb' in file]
    embs = [emb(path + file) for file in filelist]
    return embs
```

```python
embs = collect_all_embs()
print(embs[0].data[:, 1:].shape)
```

```python
data_dict = {}

for index in range(5):
    
    data = []
    
    for emb in embs:
        projected = PCA(n_components=2).fit_transform(emb.data[:, 1:])
        data.append(list(projected[index]))

    data_dict[index] = np.array(data)
```

```python
fig, axes = plt.subplots(ncols=5, nrows=1, sharex=True, sharey=True, figsize=(5*3,3))
axes[0].set_ylabel('PC2')
    
for i in range(5):
    axes[i].set_title('Node {}'.format(i))
    axes[i].set_xlabel('PC1')
    
    axes[i].set_aspect('equal')
    axes[i].set_ylim(-2,2)
    axes[i].set_xlim(-2,2)
    axes[i].scatter(x=data_dict[i][:,0], y=data_dict[i][:,1])

    
plt.savefig('../../node_slices', dpi=500)
```
