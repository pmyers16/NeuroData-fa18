```python
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.decomposition import PCA
# from sklearn.manifold import TSNE
```

```python
hnu = pd.read_csv('data/hnu1-klocality_with_age-sex.csv')
hbn = pd.read_csv('data/hbn-klocality_with_age-sex.csv')
```

```python
pca = PCA(n_components=5)
X = pca.fit_transform(hbn[hbn.columns[3:]])
print(pca.explained_variance_ratio_)


ax = plt.axes(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=hbn['sex'])
plt.show()
```

```python
pca = PCA(n_components=5)
X = pca.fit_transform(hnu[hnu.columns[4:]])
print(pca.explained_variance_ratio_)

ax = plt.axes(projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=hnu['sex'])
plt.show()
```
